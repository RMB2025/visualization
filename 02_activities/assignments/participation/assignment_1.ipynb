```python
# libraries
import numpy as np
from numpy import linspace
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# dataframe
df = pd.DataFrame({
'var1': np.random.normal(size=1000),
'var2': np.random.normal(loc=2, size=1000) * -1
})

# Fig size
plt.rcParams["figure.figsize"]=12,8

# plot density chart for var1
sns.kdeplot(data=df, x="var1",  fill=True, alpha=1)

# plot density chart for var2
kde = gaussian_kde(df.var2)
x_range = linspace(min(df.var2), max(df.var2), len(df.var2))

# multiply by -1 to reverse axis (mirror plot)
sns.lineplot(x=x_range*-1, y=kde(x_range) * -1, color='orange') 
plt.fill_between(x_range*-1, kde(x_range) * -1, color='orange')

# add axis names        
plt.xlabel("value of x")
plt.axhline(y=0, linestyle='-',linewidth=1, color='black')

# show the graph
plt.show()

# Code obtained from the Python Graph Gallery at https://python-graph-gallery.com/all-charts/
```

```python
# import required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

# create dataset with random values
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# create a line plot with two y variables on the same axes
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize = (5, 3))
# add a legend
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc = 'lower right',
          frameon = True, #add frame around the legend
          fontsize = 12, #change font size
          ncol = 2, #specify number of columns
          shadow = True )
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
          bbox_to_anchor =(1, 1)) # position the legend outside the plot
fig.show()
```

```python
# create a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')
```

```python
# create a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

# use the text() function to specify coordinates and text
ax.text(10, 95, "This value is important!")
fig.show()
```

```python
# create a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x, y2, label = "Person 2")
ax.legend(loc='lower right')

# modify text annotation by adding arguments
ax.text(10, 95, "This value is important!",
        ha='center',    ##alignment of text
        color = 'red',  ##modify font colour
        size = 20)      ##modify font size
fig.show()
```

```python
# make a simple, empty set of axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

# position annotation with reference to the data, the axes, or the figure dimensions
ax.text(1, 5, ". Data:(1, 5)",              transform=ax.transData      )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)",      transform=ax.transAxes      )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)",    transform=fig.transFigure   )
```

```python
# Annotate with arrows and text using the annotate function. Specify the location to point at and the location to place the text
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(facecolor='black', shrink=0.05))

fig.show()
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y2 = np.array([110,180,240,99,220])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
# arrange subplots within the figure
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                   ['ax2', 'ax3']],
                                   figsize=(7, 4))
for label, ax in someaxes.items():
    ax.text(0.5, 0.5, f'This is {label!r}',
            fontsize=14,
            ha = 'center',
            transform=ax.transAxes)
```

```python
# add data to subplots
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
# modify figure by adding large x axis titles. No layout specified.
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18)

plt.show()
```

```python
# modify figure by adding constrained layout
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7, 4),
                                    layout = "constrained")                                   
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18)

plt.show()
```

```python
# make multiple viz on one axes object
# first make our sample data
x = np.array(["Luffy", "Zoro", "nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot element to the same axes (ax)
ax.bar(x, y1, color ="indigo")
ax.plot(x, y2, color = "red")
```

```python
# add error information
# first calculate standard deviation of the data
y2_sd = np.std(y2)

# then plot the line
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, 
        color = "red")
```

```python

# use errorbar() to add in the error line
ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd, #yerr specifies that we're plotting vertical error bars
            fmt = "none", #fmt makes sure we're not plotting the actual data points, only the error
            ecolor= "indigo",
            elinewidth = 4,
            capsize = 6,
            capthick = 4 )
```
