```python
# adding images to plots

# load libraries
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images

# get image from the internet
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image  = Image.open(image_file)

# reuse data to make a basic line plot
fig, ax = plt.subplots(figsize = (7, 3))
ax.plot(x, y2, color = "red")

# overlay a new axis ('ax_image') on the figure (on top of 'ax') to act as a container for the image
ax_image = fig.add_axes ( [0.1, # x coordinate (ON FIGURE, NOT AXES)
                          0.11, # y coordinate (ON FIGURE, NOT AXES)
                          0.15, # image width
                          0.35]) # image height

# add imshow() to add the image
ax_image.imshow(image)
ax_image.axis('off')
plt.show()

```

```python
# install seaborn, then other packages
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# load tips dataset which contains information about restaurant tips
tips = sns.load_dataset("tips")
print(tips)
```

```python
# create plot of tip amount vs total bill
sns.lineplot(data=tips,  # choose our dataset
            x='total_bill', # define our x variable
            y='tip')     # define our y variable
```

```python
# use premade styles to alter the appearance of the plot
sns.lineplot(data=tips,  # choose our dataset
            x='total_bill', # define our x variable
            y='tip')     # define our y variable
sns.set_style('darkgrid') # other preset styles are 'whitegrid', 'dark', 'white', 'ticks', etc
```

```python
# use seaborn to add title and axis labels
tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip')

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
# modify aesthetic elements

# change figure size
fig = plt.subplots(figsize=(10, 3))
```

```python
fig = plt.subplots(figsize=(10, 3))

tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip',
                        color = 'hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')

tipgraph.set(title = 'Tips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
# Seaborn makes it easy to represent multiple variables with different visual elements of the graph

tipgraph = sns.scatterplot(data=tips, x='total_bill',
                           y='tip', style = 'time', 
                           hue = 'day', palette = ['purple','hotpink', 'deepskyblue', 'yellowgreen'])

tipgraph.set(title='Tips vs. Total Bill',
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
# pairplot is another convenient way to compare variables
sns.pairplot(data = tips, hue = 'day')
```

```python
# relplot also lets us explore specific levels within variables
daysplot = sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex",
    col="day",
    kind="scatter",
    col_wrap=2
)
```

Dynamic Data Visualizations

```python
# import package and create sample data
import plotly.graph_objects as go #'go' is 'graph objects'

x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
# create the plot
graph = go.Figure()

# choose plot type (bar) and data
graph.add_trace(go.Bar(x=x1, y=y1))

# update layout to include titles
graph.update_layout(title="Pirate Scores",
    xaxis_title="Pirates",
    yaxis_title="Score")

# show the plot
graph.show()

# If we hove over the bars we see automatic labels
# If we click and drag over a portion of the plot, we canzoom in. Double clicking zooms back out.
# We can click and drag on an axis to scroll.
# The upper-right menu gives options to save as PNG, pan/zoom, and box/lasso select
```

```python
# To save plotly graphs as HTML files and embed them in webpages:
# graph.write_html("../../folders/pirategraph.html")
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers', # we want points for a scatter plot
        marker=dict(
            size=15,                            # point size
            color='hotpink',                    # point colour
            opacity=1,                          # point transparency/alpha
            line=dict(width=5, color='purple')  # point outline
        )))

graph.update_layout(
    title='Interactive Pirate Plot',
    xaxis_title='Pirates',
    yaxis_title='Scores',
    width=500, height=500
)
```

```python
# import wordcloud package and load sample dataset of movie quotes

from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
on_bad_lines='skip')
df
```

```python
# join all our text from each row from our quote column into a string
text = "".join(each for each in df.quote)

# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
colormap = 'inferno').generate(text)

# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, # 'imshow' was used when adding pictures to our matplotlib axes
          interpolation='bilinear') # this line helps smooth the image
ax.axis("off")
plt.show()
```

```python
# modifying venn diagrams
```

```python
# define the sets
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])

diagram = venn2_unweighted([A, B],
                           set_labels = ('Fruits', 'Pies'),
                           set_colors=("blue", "red"),
                           alpha=0.5)

diagram.get_label_by_id("10").set_text("\n".join(A - B)) #left

diagram.get_label_by_id("11").set_text("\n".join(A & B)) #middle

diagram.get_label_by_id("01").set_text("\n".join(B - A)) #right
```
